```python
# Part 2: Decode!

#   0:      1:      2:      3:      4:
#  aaaa    ....    aaaa    aaaa    ....
# b    c  .    c  .    c  .    c  b    c
# b    c  .    c  .    c  .    c  b    c
#  ....    ....    dddd    dddd    dddd
# e    f  .    f  e    .  .    f  .    f
# e    f  .    f  e    .  .    f  .    f
#  gggg    ....    gggg    gggg    ....

#   5:      6:      7:      8:      9:
#  aaaa    aaaa    aaaa    aaaa    aaaa
# b    .  b    .  .    c  b    c  b    c
# b    .  b    .  .    c  b    c  b    c
#  dddd    dddd    ....    dddd    dddd
# .    f  e    f  .    f  e    f  .    f
# .    f  e    f  .    f  e    f  .    f
#  gggg    gggg    ....    gggg    gggg

# ---
# Segments count pairs
# # of Seg | Values
# --------------------
#    0          -
#    1          -
#    2          1
#    3          7
#    4          4
#    5          2, 3, 5
#    6          0, 6, 9
#    7          8


fpath = "input.txt"
# fpath = "test.txt"
# fpath = "test1.txt"
with open(fpath, 'r') as ifp:
    data = []
    for line in ifp.readlines():
        line = line.strip()
        # print(f"{line=}")
        patterns, digits = line.split("|")
        patterns = patterns.strip().split(" ")
        digits = digits.strip().split(" ")
        data.append((patterns, digits))
data

from pprint import pprint as print


total = 0
for line in data:
    patterns, digits = line
    # print(patterns)
    
    try:
        # Identify unique ones
        codes = {}
        for p in patterns:
            if len(p) == 4:
                codes[4] = ''.join(set(p))
            elif len(p) == 3:
                codes[7] = p
            elif len(p) == 7:
                codes[8] = p
        
        # Case: 1
        # NOTE:at least one test case does not provide 1
        # Logic: if you remove the segments of '4' from '7', you get top segment, and if you
        #        remove the top segment from '7', you get '1'
        four_code = set(codes[4])
        seven_code = set(codes[7])
        one_code = seven_code - (seven_code - four_code)
        codes[1] = ''.join(one_code)
        
        # Case: 5 segment lengths: 2, 3, & 5
        # Case: 2
        # Logic: if you remove the segments of '2', '3', or '5' from '4', '2' is unique with two segments left.
        for code in [ p for p in patterns if len(p) == 5 ]:
            unknown_code = set(code)
            four_code = set(codes[4])
            if len((four_code - unknown_code)) == 2:
                # print(f"Found '2': {code=}")
                codes[2] = code
                
        # Case: 3 & 5
        # Logic: if you remove the segments from '7' from '3', two segments remain
        #        if you remove the segments from '7' from '5', three segments remain
        for code in [ p for p in patterns if len(p) == 5 and p != codes[2] ]:
            unknown_code = set(code)
            seven_code = set(codes[7])
            if len(unknown_code - seven_code) == 2:
                # print(f"Found '3': {code=}")
                codes[3] = code
            else:
                # print(f"Found '5': {code=}")
                codes[5] = code
            
        # Case 6 segment lengths: 0, 6, & 9
        # Case: 6
        # Logic: if you remove the segments of '7' from '6', 4 segments remain
        #        whereas for '0' and '9' have 3 left over.
        for code in [ p for p in patterns if len(p) == 6 ]:
            unknown_code = set(code)
            seven_code = set(codes[7])
            if len(unknown_code - seven_code) == 4:
                # print(f"Found '6': {code=}")
                codes[6] = code
        
        # Case: 0 & 9
        # Logic: if you remove the segments of '5' from '0', 2 segments remain
        #        if you remove the segments of '5' from '9', 3 segments remain
        for code in [ p for p in patterns if len(p) == 6 and p != codes[6] ]:
            unknown_code = set(code)
            five_code = set(codes[5])
            if len(unknown_code - five_code) == 1:
                # print(f"Found '9': {code=}")
                codes[9] = code
            else:
                # print(f"Found '0': {code=}")
                codes[0] = code
        
        # Swap key-value pairs
        # print(f"{codes=}")
        codes = dict([(''.join(sorted(list(value))), key) for key, value in codes.items()])
        # print(codes)
        
        # Decode line
        decoded_digits = []
        for digit in digits:
            setified_digits = ''.join(sorted(list(digit)))
            decoded_digits.append(codes[setified_digits])
    except Exception as e:
        raise
    else:
        # Convert to int
        value = int(''.join([ str(d) for d in decoded_digits ]))
        # print(value)
        total += value

    
print(f"{total=}")
    
```
