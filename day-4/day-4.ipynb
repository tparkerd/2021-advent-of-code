```python
import numpy as np
import pandas as pd

fpath = 'input.txt'
# fpath = 'test.txt'
with open(fpath, 'r') as ifp:
    called_numbers = ifp.readline()
    called_numbers = list(map(int, called_numbers.split(',')))
    print(f"{called_numbers=}")
    
    # Read remaining lines
    lines = [l.strip() for l in ifp.readlines()]
    print(f"{lines=}")
    
boards = []
board_row_idx = 0

for idx, line in enumerate(lines, start=1):
    if line == '':
        board = lines[idx:idx+5]
        boards.append(board)

# Convert to numeric lists
for idx, board in enumerate(boards):
    for rdx, row in enumerate(board):
        row = list(map(int, row.split()))
        board[rdx] = row
    arr = np.array(boards[idx], dtype=np.uint8)
    boards[idx] = arr
print(f"{boards=}")
```

```python
def solve(numbers, boards):
    # Set up state boards
    state = []
    for idx, board in enumerate(boards):
        state.append(np.zeros(board.shape, dtype=np.uint8))
        
    def sumUnusedSlots(board, board_state):
        return np.sum(board[np.where(board_state == 0)])

    # For each called number, progress the state of the game
    for n in numbers:
        for idx, board in enumerate(boards):
            # For each board, update its state and check if it's complete
            
            # Find coordinates of called number
            coords = np.where(board == n)
            state[idx][coords] = 1
            x, y = coords
            
            # Check for bingo for just the row-column that was updated
            # Row-wise
            row_sum = np.sum(state[idx][x,:])
            # Column-wise
            col_sum = np.sum(state[idx][:,y])
            if row_sum >= 5 or col_sum >= 5:
                print("Bingo!")
                return sumUnusedSlots(board, state[idx]) * n

result = solve(called_numbers, boards)
print(f"{result=}")
```

```python
# Part 2: Pick the last-to-win board to guarantee that the opponent wins

def solve_part2(numbers, boards):
    # Set up state boards
    state = []
    for idx, board in enumerate(boards):
        state.append(np.zeros(board.shape, dtype=np.uint8))
        
    def sumUnusedSlots(board, board_state):
        return np.sum(board[np.where(board_state == 0)])

    # For each called number, progress the state of the game
    result = None
    last_winner_index = -1
    winner_boards_order = [ None for _ in boards ]
    for n in numbers:
        for idx, board in enumerate(boards):
            # Base case: If it's already won, ignore the board
            if winner_boards_order[idx] is not None:
                continue
            
            # For each board, update its state and check if it's complete
            
            # Find coordinates of called number
            coords = np.where(board == n)
            state[idx][coords] = 1
            x, y = coords
            
            # Check for bingo for just the row-column that was updated
            # Row-wise
            row_sum = np.sum(state[idx][x,:])
            # Column-wise
            col_sum = np.sum(state[idx][:,y])
            if row_sum >= 5 or col_sum >= 5:
                # result = sumUnusedSlots(board, state[idx]) * n
                winner_boards_order[idx] = sumUnusedSlots(board, state[idx]) * n
                last_winner_index = idx
               
     
    return winner_boards_order[last_winner_index]

result = solve_part2(called_numbers, boards)
print(f"{result=}")
```
