```python
# Base solver
def solve(days, state): 
    for day in range(1, days+1):
        # Process current fish
        adult_fish = len(state)
        for idx in range(adult_fish):
            if state[idx] == 0:
                state[idx] = 6
                state.append(8)
            else:
                state[idx] -= 1

    return days, adult_fish, state
```

```python
# Part 2!
days = 256

# Initalize known values
m = np.zeros(shape=(days+2, 9))
m = pd.DataFrame(m)
m.index.name = 'day'
for day in range(1,9):
    for age in m.columns:
        _, fish_count, _ = solve(day, [age])
        m.iloc[day, age] = fish_count

m.iloc[0, 0:8] = 1
m.iloc[0:2,0] = 2

# Dynamically solve the remaining values
for day in range(9, days + 2):
    # For each of the actual 8 ages to consider, calculate the value
    for age in range(0, 8):
        # print(f"{day=}, {age=}")
        m.iloc[day, age] = m.iloc[day-7, age] + m.iloc[day-8, age+1]
    # Literal edge case, the 8th age
    m.iloc[day, 8] = m.iloc[day-1, 7]
m.to_csv("result.csv")
m
```
