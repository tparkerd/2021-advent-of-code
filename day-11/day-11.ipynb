```python
# Part 1! Flashing dumbo octopuses!
import numpy as np

fpath = 'input.txt'
# fpath = 'test.txt'
# fpath = 'test2.txt'
# fpath = 'test3.txt'

matrix = np.genfromtxt(fpath, delimiter=1, dtype=np.uint8)

def resolve(m, state=None):
    """Resolve flashing"""
    height, width = m.shape
    if state is None:
        pts = np.argwhere(m == 10)
    else:
        pts = state
    
    # Base case: no new flashes to resolve
    if len(pts) == 0:
        return m

    flashing_pts = []
    # Otherwise, resolve flashing
    for pt in pts:
        y, x = pt
        offsets = [-1, 0, 1]
        for offy in offsets:
            for offx in offsets:
                # Skip self
                # if offx == 0 and offy == 0:
                #     continue
                
                ny = offy + y
                nx = offx + x
                # Keep in bounds
                if (ny < 0 or ny >= height) or (nx < 0 or nx >= width):
                    continue
                else:
                    m[(ny, nx)] += 1
                    if m[(ny, nx)] == 10:
                        flashing_pts.append((ny, nx))
                    
    return resolve(m, flashing_pts)

def take_a_step(m):
    # Increment energy level for all octopuses
    m += 1
    
    # Let the octopuses flash!
    m = resolve(m)

    # Reset energy levels of octopuses that flashed
    m[m>9] = 0
    flashes = np.argwhere(m == 0)

    return m, len(flashes)

def solve(m, steps=100):
    total = 0
    print(f"Before any steps:")
    print(m, end="\n\n")
    for i in range(steps):
        m, flash_counts = take_a_step(m)
        total += flash_counts
        print(f"After step {i+1}:")
        for row in m:
            print(row)
        print(f"----------------", end="\n\n")
    return total

# solve(matrix, 10)
solve(matrix, 100)
```
