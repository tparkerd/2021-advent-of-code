```python
fpath = 'input.txt'
with open(fpath, 'r') as ifp:
    cIncrease = 0
    cDecrease = 0
    cEqual = 0
    cLines = 0
    prev_value = None
    for line in ifp.readlines():
        cLines += 1
        line = int(line)
        if prev_value is not None:
            if line > prev_value:
                cIncrease += 1
            elif line < prev_value:
                cDecrease += 1
            else:
                cEqual += 1

        # Set current value to previous before proceeding to next line
        prev_value = line

print(f"{cIncrease=}")
print(f"{cDecrease=}")
print(f"{cEqual=}")
print(f"{cLines=}")
```

```python
with open(fpath, 'r') as ifp:
    data = []
    lines = ifp.readlines()
    for line in lines:
        data.append(int(line))

# Loop over with a window
# Initialize with first window
prev_value = sum(data[:3])
cIncrease = 0
cDecrease = 0
cEqual = 0
cWindow = 1
# Start the window on the second element and end when there are only
# three elements left
for idx in range(1, len(data)-2):
    current_value = sum(data[idx:idx+3])
    if current_value > prev_value:
        cIncrease += 1
    elif current_value < prev_value:
        cDecrease += 1
    else:
        cEqual += 1

    prev_value = current_value
    
print(f"{cIncrease=}")
print(f"{cDecrease=}")
print(f"{cEqual=}")
```
